Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
data = pd.read_csv('US_Accidents_March23.csv')
```

```python
print(data.info())
```

```python
print(data.head())
```

```python
# Visualizations
# Plotting accident severity distribution
plt.figure(figsize=(8, 6))
sns.countplot(data['Severity'])
plt.title('Accident Severity Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting accidents by weather condition
plt.figure(figsize=(12, 6))
sns.countplot(x='Weather_Condition', data=data, order=data['Weather_Condition'].value_counts().index[:10], palette='viridis')
plt.title('Accidents by Weather Condition (Top 10)')
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plotting accident hotspots using latitude and longitude
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=data, hue='Severity', palette='viridis', alpha=0.5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Severity')
plt.show()
```

```python
# Identifying contributing factors
factors = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 
           'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=factors, y=data[factors].sum(), palette='viridis')
plt.title('Contributing Factors to Accidents')
plt.xlabel('Factor')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
